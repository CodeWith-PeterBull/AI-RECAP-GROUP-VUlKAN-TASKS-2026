This notebook demonstrates **data understanding and data indexing using Python (pandas)** based on the Netflix Titles Dataset.

```python
# Import necessary libraries for data analysis

import pandas as pd
import numpy as np

# pandas is used for data manipulation and analysis
# numpy supports numerical operations
```

```python
# Load the dataset into a pandas DataFrame

df = pd.read_csv("netflix_titles.csv")

# Display the first few rows of the dataset to confirm it loaded correctly
df.head()
```

## Data Understanding
Question 1
How many rows and columns are contained in the dataset?

```python
# The shape attribute returns a tuple: (rows, columns)

df.shape

# Extract rows and columns separately for clarity

rows, columns = df.shape

print("Number of rows:", rows)
print("Number of columns:", columns)
```

Question 2
What datatypes are contained in the dataset?

```python
# df.info() shows column names, non-null counts and data types

df.info()

# We can also list just the data types

df.dtypes
```

Question 3
Is the dataset complete (i.e., are there missing values)?

```python
# Check missing values for each column

df.isnull().sum()

# If any column has values greater than 0, it means missing values exist

missing_values = df.isnull().sum()

print("Missing values per column:")
print(missing_values)
```

Question 4
Among the last 20 rows, how many are Movies and how many are TV Shows?

```python
# Retrieve the last 20 rows of the dataset

last_20_rows = df.tail(20)

# Display them
last_20_rows

# Count the number of Movies and TV Shows in these rows

last_20_rows['type'].value_counts()

# Store the result for clearer output

type_count = last_20_rows['type'].value_counts()

print(type_count)
```

## Data Indexing
   
Question 1: 
Give the number of all Movies originating from South Africa

```python
# Filter rows where:
# type = Movie
# country contains South Africa

south_africa_movies = df[
    (df['type'] == 'Movie') &
    (df['country'].str.contains("South Africa", na=False))
]

# Count the number of such movies

len(south_africa_movies)

print("Number of Movies from South Africa:", len(south_africa_movies))
```

Question 2
What is the description of the movie in the 12th row?

```python
# Using iloc to access rows by index
# Note: Python indexing starts at 0
# So row 12 corresponds to index 11

df.iloc[11]

# Extract only the description column

description_12th = df.iloc[11]['description']

print("Description of movie in the 12th row:")
print(description_12th)
```

Question 3
From the South African movies, list only those released in 2021

```python
# Filter South African movies released in 2021

south_africa_2021 = df[
    (df['type'] == 'Movie') &
    (df['country'].str.contains("South Africa", na=False)) &
    (df['release_year'] == 2021)
]

south_africa_2021

# Display selected columns for clarity

south_africa_2021[['title', 'release_year', 'country']]
```

Question 4
List the entire cast of all movies released in 2021

```python
# Filter movies released in 2021

movies_2021 = df[
    (df['type'] == 'Movie') &
    (df['release_year'] == 2021)
]

# Display the title and cast

movies_2021[['title', 'cast']]

# Extract just the cast column

cast_2021 = movies_2021['cast']

cast_2021
```
